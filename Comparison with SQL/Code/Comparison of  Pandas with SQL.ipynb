````python
import numpy as np
import pandas as pd
````

````python
Tips = pd.read_csv('../Data/Tips.csv')
Tips.rename(columns = {'total_bill':'Bill', 
                      'tip':'Tip', 
                      'sex':'Sex', 
                      'smoker':'Smoke', 
                      'day':'Day', 
                      'time':'Meal', 
                      'size':'Size'}, 
           inplace = True)
Tips.head()
````

**SELECT**

```SQL
SELECT Bill, Tip, Smoke, Meal
FROM Tips
LIMIT 5;
```

````python
Tips[['Bill','Tip','Smoke','Meal']].head()
````

**SELECT** and **CALCULATE**
```SQL
SELECT *, Tip / Bill as Rate
FROM Tips
LIMIT 5;
```

DataFrame.**assign**(New Column) : Append a New Column.

````python
Tips.assign(Rate = Tips['Tip'] / Tips['Bill']).head()
````

**NULL** Checking using **notna()** and **isna()**

````python
Frame = pd.DataFrame(
    {'Col 1':['A', 'B', np.NAN, 'C', 'D'], 
     'Col 2':['F', np.NAN, 'G', 'H', 'I']}
)
Frame
````

**SELECT IS NULL**

```SQL
SELECT * 
FROM Frame
WHERE 'Col 2' IS NULL;
```

**isna( )**

````python
Frame[Frame['Col 2'].isna()]
````

**SELECT IS NOT NULL**

```SQL
SELECT * 
FROM Frame

```
**notna( )**

````python
Frame[Frame['Col 2'].notna()]
````

**GROUP BY**

```SQL
SELECT Sex, Count(*)
FROM Tips
GROUP BY Sex;
```

````python
Tips.groupby('Sex').size()
````

Using **size( )** is Better than **count( )**

````python
Tips.groupby('Sex').count()
````

````python
Tips.groupby('Sex')['Bill'].count()
````

**GROUP BY AGGREGATION**

```SQL
SELECT Day, AVG(Tip), COUNT(*)
FROM Tips
GROUP BY Day
```

````python
Tips.groupby('Day').agg({'Tip':np.mean, 'Day':np.size})
````

**GROUP BY** More than One Column 

```SQL
SELECT Smoke, Day, COUNT(*), AVG(Tip)
FROM Tips
GROUP BY Smoke, Day;
```

````python
Tips.groupby(['Smoke','Day']).agg({'Tip':[np.size, np.mean]})
````

**JOIN**

Join can be Performed with **join( )** or **merge( )**

**Types** : ( LEFT, RIGHT, INNER, FULL )

````python
df1 = pd.DataFrame({'Key':['A','B','C','D'], 
                    'Value':np.random.randn(4)})

df2 = pd.DataFrame({'Key':['B','D','D','E'], 
                    'Value':np.random.randn(4)})
````

````python
df1
````

````python
df2
````

**INNER JOIN**

```SQL
SELECT * 
FROM df1
INNER JOIN df2
ON df1.Key = df2.Key
```

**merge( )** offers Better Parameters to Perform Join Operations.

Default : **Inner** Join

````python
pd.merge(left = df1, right = df2, on = 'Key') 
````

**LEFT OUTER JOIN**

```SQL
SELECT * 
FROM df1
LEFT OUTER JOIN df2
ON df1.Key = df2.Key;
```

**how** = **"left"**

````python
pd.merge(left = df1, right = df2, on = 'Key', how = 'left')
````

**RIGHT OUTER JOIN**

```SQL
SELECT * 
FROM df1
RIGHT OUTER JOIN df2
ON df1.Key = df2.Key;
```

**how** = **"right"**

````python
pd.merge(left = df1, right = df2, on = 'Key', how = 'right')
````

**FULL OUTER JOIN**

```SQL
SELECT *
FROM df1
FULL OUTER JOIN df2
ON df1.Key = df2.Key;
```
**how** = **"outer"**

````python
pd.merge(left = df1, right = df2, on = 'Key', how = 'outer')
````

````python
df1 = pd.DataFrame(
    {'City':['Noida','Banglore','Hyderabad'], 
     'Rank':range(1,4)}
)

df2 = pd.DataFrame(
    {'City':['Noida','Gurugram','Pune'], 
     'Rank':[1,4,5]}
)
````

````python
df1
````

````python
df2
````

**UNION ALL**

Including Duplicates.

```SQL
SELECT City, Rank
FROM df1
UNION ALL
SELECT City, Rank
FROM df2;
```

Union All can be Performed using **concat( )**

````python
pd.concat([df1, df2])
````

**UNION**

No Duplicates

```SQL
SELECT City, Rank
FROM df1
UNION
SELECT City, Rank
FROM df2
```

**concat( )** with **drop_duplicates( )**

````python
pd.concat([df1, df2]).drop_duplicates()
````
